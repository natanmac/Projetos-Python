```python
# media final
# 4 notas
# 10 alunos

# poder mudar somente uma nota de um aluno
# usar conceito de listas

# colocar se está reprovado ou aprovado

import matplotlib.pyplot as plt

listaDeAlunos = [{'Nome': 'natan', "Nota 1": 0.0, "Nota 2": 0.0, "Nota 3": 0.0, "Nota 4": 0.0, "Média": 0.0}]

def menuPrincipal():
    print("MENU PRINCIPAL")
    print("1 - Inserir um novo aluno")
    print("2 - Alterar um aluno ou nota da lista")
    print("3 - Consultar lista")
    print("4 - Plotar gráfico de um aluno")
    print("5 - Finalizar código")
    valorMenu = float(input("Escolha uma das opções: "))


    if(valorMenu == 1):
        print("-" * 50)
        inserirAluno()
    elif(valorMenu == 2):
        print("-" * 50)
        alterarNomeNota()
    elif(valorMenu == 3):
        print("-" * 50)
        consultarTabela()
    elif(valorMenu == 4):
        print("-" * 50)
        plotarGrafico()
    elif(valorMenu == 5):
        print("-" * 50)
        encerrar()
    else:
        print("Inválido, insira outro valor")
        print("-" * 50)
        menuPrincipal()



def inserirAluno():
    while True:
        try:
            nome = input("Digite o nome do aluno: ")
            N1 = float(input("Digite a primeira nota: "))
            N2 = float(input("Digite a segunda nota: "))
            N3 = float(input("Digite a terceira nota: "))
            N4 = float(input("Digite a quarta nota: "))
            media = (N1 + N2 + N3 + N4)/4
            aluno = {'Nome': nome, "Nota 1": N1, "Nota 2": N2, "Nota 3": N3, "Nota 4": N4, "Média": media}
            listaDeAlunos.append(aluno)
            print(listaDeAlunos)
            print("Novo aluno inserido!")
            print("-" * 50)
            return listaDeAlunos, menuPrincipal()
        except ValueError:
            print("Valor digitado incorreto: ")
        

def consultarTabela():
    print("1 - Consultar todos da lista")
    print("2 - Consultar por nome")
    print("3 - Voltar")
    consulta = float(input("Escolha uma opção: "))

    if(consulta == 1):
        for n in listaDeAlunos:
            print(n)
    elif(consulta == 2):               
        nomeConsulta = str(input("Digite o nome do aluno a ser consultado: "))
        for n in listaDeAlunos:
            ix = listaDeAlunos.index(n)
            if(nomeConsulta == listaDeAlunos[ix]['Nome']):
                nomeEncontrado = n
                print(nomeEncontrado)
            else:
                pass
    elif(consulta == 3):
        menuPrincipal()
    else:
        print("Valor digitado inválido")
        consultarTabela()
    menuPrincipal()

def alterarNomeNota():
    print("1 - Apagar um aluno")
    print("2 - Alterar nota de um aluno")
    print("3 - Voltar")
    consultaNomeNota = float(input("Escolha uma opção: "))

    if(consultaNomeNota == 1):
        nomeApagar = str(input("Digite o nome do aluno: "))
        for i in listaDeAlunos:
            iApagar = listaDeAlunos.index(i)
            if(nomeApagar == listaDeAlunos[iApagar]['Nome']):
                nomeDelete = i
                certeza = input("Tem certeza que deseja apagar esse aluno? (s/n)")
                if(certeza == "s"):
                    listaDeAlunos.remove(nomeDelete)
                    print("Nome apagado")
                else:
                    print("O nome não foi apagado!")
                    menuPrincipal()
            else:
                print("Nome não encontrado")
                menuPrincipal()
    elif(consultaNomeNota == 2):
        alterarNota = str(input("Digite o nome do aluno que deseja alterar a nota: "))
        for i in listaDeAlunos:
            ix = listaDeAlunos.index(i)
            if(alterarNota == listaDeAlunos[ix]['Nome']):
                print("Aluno encontrado: ")
                print(i)
                valorAltNota = int(input(("Qual nota quer alterar: 1 - N1; 2 - N2, 3 - N3; 4 - N4: ")))
                if(valorAltNota == 1):
                    print("O valor atual da nota é: ", listaDeAlunos[ix]['Nota 1'])
                    novaNota = float(input("Digite o valor da nova nota: "))
                    listaDeAlunos[ix]['Nota 1'] = novaNota
                    media = (listaDeAlunos[ix]['Nota 1'] + listaDeAlunos[ix]['Nota 2'] + 
                    listaDeAlunos[ix]['Nota 3'] + listaDeAlunos[ix]['Nota 4'])/4
                    listaDeAlunos[ix]['Média'] = media
                    print("Valor alterado com sucesso!")
                elif(valorAltNota == 2):
                    print("O valor atual da nota é: ", listaDeAlunos[ix]['Nota 2'])
                    novaNota = float(input("Digite o valor da nova nota: "))
                    listaDeAlunos[ix]['Nota 2'] = novaNota
                    media = (listaDeAlunos[ix]['Nota 1'] + listaDeAlunos[ix]['Nota 2'] + 
                    listaDeAlunos[ix]['Nota 3'] + listaDeAlunos[ix]['Nota 4'])/4
                    listaDeAlunos[ix]['Média'] = media
                    print("Valor alterado com sucesso!")
                elif(valorAltNota == 3):
                    print("O valor atual da nota é: ", listaDeAlunos[ix]['Nota 3'])
                    novaNota = float(input("Digite o valor da nova nota: "))
                    listaDeAlunos[ix]['Nota 3'] = novaNota
                    media = (listaDeAlunos[ix]['Nota 1'] + listaDeAlunos[ix]['Nota 2'] + 
                    listaDeAlunos[ix]['Nota 3'] + listaDeAlunos[ix]['Nota 4'])/4
                    listaDeAlunos[ix]['Média'] = media
                    print("Valor alterado com sucesso!")
                elif(valorAltNota == 4):
                    print("O valor atual da nota é: ", listaDeAlunos[ix]['Nota 4'])
                    novaNota = float(input("Digite o valor da nova nota: "))
                    listaDeAlunos[ix]['Nota 4'] = novaNota
                    media = (listaDeAlunos[ix]['Nota 1'] + listaDeAlunos[ix]['Nota 2'] + 
                    listaDeAlunos[ix]['Nota 3'] + listaDeAlunos[ix]['Nota 4'])/4
                    listaDeAlunos[ix]['Média'] = media
                    print("Valor alterado com sucesso!")
                else:
                    print("Valor não encontrado")
                    menuPrincipal()
    else:
        pass
    menuPrincipal()

def plotarGrafico():
    graficoAluno = str(input("Digite o nome do aluno que deseja plotar gráfico: "))
    for i in listaDeAlunos:
        ix = listaDeAlunos.index(i)
        if(graficoAluno == listaDeAlunos[ix]['Nome']):
            plt.plot(["1ºbim", "2ºbim", "3ºbim", "4ºbim"], [listaDeAlunos[ix]['Nota 1'], listaDeAlunos[ix]['Nota 2'], 
            listaDeAlunos[ix]['Nota 3'], listaDeAlunos[ix]['Nota 4']])
            print("Gráfico do aluno!")
            plt.show()
            menuPrincipal()
        else:
            print("Nome não encontrado!")
    menuPrincipal()

menuPrincipal()

def encerrar():
    print("Código encerrado")
    exit()
```
